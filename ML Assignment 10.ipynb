Unsupervised learning is a type of machine learning where the model is trained on data that does not have labeled outputs.

In simple terms :

```python
'''Q.2  How does K-Means clustering algorithm work?'''
```

### How K-Means Clustering Algorithm Works (Exam View):

K-Means is an unsupervised learning algorithm used for clustering data into K distinct groups based on similarity.

Steps:

1. Choose the number of clusters (K)
   Decide how many clusters you want to form.

2. Initialize K centroids
   Randomly select K points from the dataset as the initial cluster centers (centroids).

3. Assign data points to the nearest centroid
   For each data point, calculate the distance to each centroid and assign it to the nearest one.

4. Update centroids
   Recalculate the centroid of each cluster as the mean of all points assigned to it.

5. Repeat steps 3 and 4
   Continue until the centroids stop changing (convergence) or a maximum number of iterations is reached.

Key Concepts:

* Centroid: The center point of a cluster
* Distance metric: Usually Euclidean distance is used
* Objective: Minimize the within-cluster sum of squares (WCSS)

Example:

If we set K = 3 for customer purchase data, K-Means will group similar purchase patterns into 3 clusters.




```python
'''Q.3  Explain the concept of a dendrogram in hierarchical clustering'''
```



A dendrogram is a tree-like diagram that shows the arrangement of clusters formed by hierarchical clustering.

Key Points:

* It displays how data points are merged or split step-by-step.
* Each leaf represents an individual data point.
* Branches show how clusters are combined.
* The height of the branches (vertical axis) represents the distance or dissimilarity between clusters.

How to Use It:

* Start from the bottom where each data point is its own cluster.
* Move upward as clusters are merged.
* To form final clusters, cut the dendrogram horizontally at a chosen level. The number of vertical lines it intersects equals the number of clusters.

Example:

In a dendrogram for 10 students based on exam scores, students with similar scores are merged early (lower height), and different groups appear clearly when the tree is cut at a certain level.

```python
'''Q.5 What are the advantages of DBSCAN over K-Means?'''
```



1. No need to specify the number of clusters (K)
   DBSCAN automatically determines the number of clusters based on data density.

2. Can detect arbitrarily shaped clusters
   Unlike K-Means which only finds spherical clusters, DBSCAN can find complex shapes.

3. Can identify outliers (noise)
   DBSCAN labels sparse data points that don't belong to any cluster as noise, which K-Means cannot do.

4. Works well with clusters of different sizes and densities
   DBSCAN can handle clusters that vary in density and size, which K-Means struggles with.

5. Deterministic results
   DBSCAN produces the same output each time, while K-Means may give different results due to random initialization.

Summary:

DBSCAN is more flexible and robust in handling noisy, non-spherical, and irregularly shaped clusters, whereas K-Means works best with well-separated and equal-sized spherical clusters.



The Silhouette Score is used to evaluate the quality of clustering by measuring how well each data point fits within its cluster compared to other clusters.

When to Use:

1. To choose the optimal number of clusters (K)
   Helps in deciding the best K in algorithms like K-Means.

2. To compare clustering algorithms
   Useful for comparing models like K-Means, DBSCAN, and Hierarchical Clustering.

3. To check how well-separated clusters are
   A higher score indicates well-separated and cohesive clusters.

Value Range:

* Ranges from -1 to 1:

  * +1 → well-clustered
  * 0 → overlapping clusters
  * −1 → wrong cluster assignment

Summary:

Use Silhouette Score to validate clustering performance, especially when labels are unknown (unsupervised learning). It helps in assessing compactness and separation of clusters.


```python
'''Q.7  What are the limitations of Hierarchical Clustering?'''
```



1. Scalability Issues

   * Not suitable for large datasets
   * Time complexity is O(n²), which makes it slow for thousands of data points

2. No Reassignment

   * Once a data point is merged into a cluster, it cannot be moved
   * May lead to irreversible mistakes

3. Choice of Linkage Affects Result

   * Different linkage methods (single, complete, average) can produce very different clusters

4. Sensitive to Noise and Outliers

   * Can be easily affected by outliers, distorting the cluster structure

5. Hard to Choose Optimal Number of Clusters

   * Requires visual inspection of the dendrogram or other metrics to decide where to cut

Summary:

Hierarchical clustering is useful for small datasets and visualizing cluster relationships, but it has performance, flexibility, and sensitivity issues on large or noisy data.

Reasons:

1. K-Means is distance-based

   * It uses Euclidean distance to assign points to clusters
   * If features are on different scales, variables with larger ranges dominate the distance, leading to biased clustering

2. Improves accuracy

   * Ensures fair contribution from all features
   * Helps form more meaningful clusters

   * Prevents features with large values (e.g., income) from overshadowing smaller-scale features (e.g., age)

* Standardization (Z-score): Mean = 0, SD = 1

Summary:

Feature scaling is essential in K-Means to avoid biased clustering results and ensure all features are treated equally during distance computation.


```python
'''Q.9  How does DBSCAN identify noise points?'''
```



DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points as data points that do not belong to any cluster based on density criteria.

Steps:

1. DBSCAN uses two parameters:

   * ε (epsilon): Neighborhood radius
   * minPts: Minimum number of points required to form a dense region

2. Point types in DBSCAN:

   * Core Point: Has at least `minPts` within its ε-radius
   * Border Point: Has fewer than `minPts` in ε, but lies in the ε of a core point
   * Noise Point (Outlier):

     * Not a core point
     * Not a border point
     * Does not belong to any cluster

3. Noise points are labeled as -1 in the cluster output.

Summary:

DBSCAN identifies noise points as data points that are too far from any dense region and don’t satisfy the minimum neighborhood density required to be part of a cluster.


```python
'''Q.10  Define inertia in the context of K-Means'''
```



Inertia is a measure of how internally coherent clusters are in K-Means clustering.

Definition:

Inertia is the sum of squared distances between each data point and the centroid of its assigned cluster.

$$
\text{Inertia} = \sum_{i=1}^{n} \|x_i - c_i\|^2
$$

Where:

* $x_i$ = data point
* $c_i$ = centroid of the cluster assigned to $x_i$

Key Points:

* Lower inertia means tighter, more compact clusters
* Used to evaluate the performance of K-Means
* Helps in choosing the optimal number of clusters using the Elbow Method

Summary:

Inertia in K-Means represents the compactness of clusters. A lower value indicates better clustering, but too low may lead to overfitting.


```python
'''Q.11  What is the elbow method in K-Means clustering?'''
```



The Elbow Method is a technique used to determine the optimal number of clusters (K) in K-Means clustering.

How It Works:

1. Run K-Means for a range of K values (e.g., 1 to 10).
2. For each K, calculate the inertia (within-cluster sum of squares).
3. Plot K vs. Inertia.
4. Look for the "elbow point" on the graph — the point where the rate of decrease sharply slows down.

Interpretation:

* The elbow point represents the optimal number of clusters, beyond which adding more clusters gives diminishing returns in reducing inertia.

Summary:

The Elbow Method helps select the best K by identifying where increasing clusters no longer significantly reduces clustering error (inertia).


```python
'''Q.12  Describe the concept of "density" in DBSCAN'''
```



In DBSCAN, density refers to how closely packed data points are in a region.

Key Concepts:

1. ε (epsilon):

   * The radius to search around a point
   * Defines the neighborhood size

2. minPts:

   * The minimum number of points required to form a dense region

3. Density Criteria:

   * A region is dense if there are at least `minPts` within the ε-radius of a point
   * Such a point is called a core point

Types of Points Based on Density:

* Core Point: Has ≥ `minPts` in its ε-neighborhood (dense region)
* Border Point: Lies within ε of a core point but has < `minPts` itself
* Noise Point: Lies in a sparse region (not dense)

Summary:

In DBSCAN, density means having a sufficient number of points within a given distance (ε). It is the basis for forming clusters without requiring predefined cluster counts.


```python
'''Q.13 Can hierarchical clustering be used on categorical data'''
```


Yes, hierarchical clustering can be used on categorical data, but it requires special distance measures suited for categorical variables.

Key Points:

1. Standard Euclidean distance is not suitable for categorical data

   * Categorical variables don't have a natural numerical order

2. Use appropriate distance metrics, such as:

   * Hamming distance: Counts mismatches between categorical values
   * Jaccard similarity: For binary attributes
   * Gower distance: Handles mixed data types (numerical + categorical)

3. Use linkage methods (like complete, average, or single) after computing the distance matrix.

Summary:

Hierarchical clustering can be applied to categorical data by using appropriate distance measures like Hamming or Gower distance. It enables clustering without converting categories into numerical values.

Key Points:

1. Silhouette Score range:

   $$
   -1 \leq \text{Silhouette Score} \leq 1
   $$

2. Negative value means:

   * The distance to the nearest neighboring cluster is less than the distance to its own cluster
   * The point is closer to another cluster than to the one it is assigned to

3. Indicates poor clustering performance and possible misclassification

Summary:

A negative Silhouette Score suggests that the data point is badly clustered and might belong to a different cluster. It is a sign of overlapping or incorrect clustering.


```python
'''Q.15 Explain the term "linkage criteria" in hierarchical clustering'''
```



Linkage criteria determine how the distance between clusters is calculated when merging them in hierarchical clustering.

Common Linkage Methods:

1. Single Linkage

   * Distance between the closest pair of points (one from each cluster)

2. Complete Linkage

   * Distance between the farthest pair of points

3. Average Linkage

   * Average distance between all pairs of points (one from each cluster)

4. Ward’s Linkage

   * Minimizes the variance within clusters; based on sum of squared differences

Summary:

Linkage criteria guide how clusters are joined in hierarchical clustering by defining the inter-cluster distance. The choice of linkage method affects the shape and size of resulting clusters.


```python
'''Q.16  Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?'''
```



K-Means clustering may perform poorly on data with uneven cluster sizes or densities because it assumes all clusters are similar in size and density.

Key Reasons:

1. Equal Variance Assumption

   * K-Means assumes clusters have similar spread (variance)
   * It struggles with compact vs. spread-out clusters

2. Centroid-Based Approach

   * Assigns points to the nearest centroid
   * May incorrectly group sparse points from a large cluster into a nearby small dense cluster

3. Insensitive to Density

   * K-Means does not consider the number of points or density in a region
   * Leads to inaccurate boundaries

4. Tendency to Form Spherical Clusters

   * Works best with circular (convex) clusters
   * Poor results on elongated or irregular shapes

Summary:

K-Means may fail on data with clusters of unequal size or density because it relies on distance to centroids and assumes uniform, spherical clusters, causing misclassification in imbalanced data.


```python
'''Q.17 What are the core parameters in DBSCAN, and how do they influence clustering?'''
```



DBSCAN (Density-Based Spatial Clustering of Applications with Noise) uses two core parameters:

1. ε (epsilon)

* Defines the radius of the neighborhood around a point
* Determines how close points must be to be considered neighbors
* Smaller ε: Detects smaller, tighter clusters
* Larger ε: Forms larger, looser clusters, may merge nearby clusters

2. minPts (minimum points)

* Minimum number of points required in an ε-radius neighborhood to form a core point
* Higher minPts: Requires denser areas to form clusters
* Lower minPts: May form clusters from sparse regions or noise

Influence on Clustering:

* Too small ε or too high minPts → Many points marked as noise
* Too large ε or too low minPts → Merges distinct clusters into one

Summary:

DBSCAN's performance depends on ε (neighborhood radius) and minPts (density threshold). Together, they define what is considered a dense region, influencing cluster shape, size, and noise detection.


```python
'''Q.18 How does K-Means++ improve upon standard K-Means initialization?'''
```



K-Means++ is an enhancement of standard K-Means that improves the initialization of cluster centroids to produce better and more stable clustering results.

Standard K-Means Initialization:

* Randomly selects k initial centroids
* Can lead to poor convergence and suboptimal clustering due to bad starting points

K-Means++ Initialization Steps:

1. Randomly choose the first centroid from the data points.
2. For each remaining point, compute its distance squared to the nearest selected centroid.
3. Select the next centroid with probability proportional to this distance squared.
4. Repeat until k centroids are chosen.

Advantages of K-Means++:

* Reduces the chances of poor clustering due to bad initialization
* Faster convergence to a better local minimum
* Improves accuracy and stability of the final clusters

Summary:

K-Means++ improves standard K-Means by smartly initializing centroids to be far apart, which often leads to better clustering performance and faster convergence.


```python
'''Q.19 What is agglomerative clustering?'''
```



Agglomerative Clustering is a type of hierarchical clustering that follows a bottom-up approach.

Key Characteristics:

* Starts with each data point as its own cluster
* Iteratively merges the two closest clusters
* Continues until all points are grouped into a single cluster or a stopping criterion is met

Steps:

1. Compute distance matrix between all data points
2. Merge the two closest clusters
3. Recompute distances between the new cluster and others (based on linkage criteria)
4. Repeat until desired number of clusters is reached

Linkage Criteria Used:

* Single Linkage (minimum distance)
* Complete Linkage (maximum distance)
* Average Linkage (mean distance)
* Ward’s Method (minimizes variance)

Summary:

Agglomerative Clustering is a bottom-up hierarchical algorithm that builds clusters by repeatedly merging the closest pairs, based on a defined linkage metric.


```python
'''Q.20 What makes Silhouette Score a better metric than just inertia for model evaluation?'''
```


Silhouette Score gives a more balanced and interpretable measure of clustering quality compared to inertia.

Inertia:

* Measures the sum of squared distances between points and their cluster centroid
* Lower inertia → tighter clusters
* But:

  * Always decreases as number of clusters increases
  * Doesn’t indicate if clusters are well-separated or overlapping

Silhouette Score:

* Measures how similar a point is to its own cluster vs. other clusters
* Ranges from -1 to 1:

  * +1: Well clustered
  * 0: On the border
  * -1: Misclassified
* Takes into account both cohesion (intra-cluster distance) and separation (inter-cluster distance)

Why It’s Better:

* More informative: Reflects how well-separated and compact the clusters are
* Helps to choose optimal number of clusters more reliably
* Not biased toward more clusters like inertia

Summary:

Silhouette Score is a better clustering evaluation metric than inertia because it considers both intra-cluster tightness and inter-cluster separation, offering a clearer picture of clustering quality.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Step 1: Generate synthetic moon-shaped data
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# Step 2: Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Step 3: Plot the results
plt.figure(figsize=(8, 6))

# Identify core clusters and outliers
unique_labels = set(labels)
colors = plt.cm.Set1.colors

for label in unique_labels:
    if label == -1:
        # Outliers (label -1)
        plt.scatter(X[labels == label][:, 0], X[labels == label][:, 1],
                    c='black', marker='x', label='Outliers')
    else:
        plt.scatter(X[labels == label][:, 0], X[labels == label][:, 1],
                    color=colors[label % len(colors)], label=f'Cluster {label}')

plt.title("DBSCAN Clustering on make_moons Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
'''Q.29 Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart'''
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Step 2: Evaluate Silhouette Scores for k = 2 to 5
silhouette_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Step 3: Plot the scores as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xticks(k_values)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different k Values (make_blobs data)")
plt.grid(axis='y')
plt.show()
```

```python
'''Q.44  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Load and scale Iris dataset
data = load_iris()
X = StandardScaler().fit_transform(data.data)

# Apply KMeans
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Calculate silhouette scores
silhouette_vals = silhouette_samples(X, cluster_labels)
avg_score = silhouette_score(X, cluster_labels)

# Plot silhouette coefficient for each sample
fig, ax = plt.subplots(figsize=(8, 6))

y_lower = 10
for i in range(k):
    ith_cluster_silhouette_vals = silhouette_vals[cluster_labels == i]
    ith_cluster_silhouette_vals.sort()

    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.Set1(i)
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                     0, ith_cluster_silhouette_vals,
                     facecolor=color, edgecolor=color, alpha=0.7)

    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, f"Cluster {i}")
    y_lower = y_upper + 10  # spacing between clusters

ax.axvline(x=avg_score, color="red", linestyle="--")
ax.set_title("Silhouette Plot for KMeans on Iris Dataset")
ax.set_xlabel("Silhouette Coefficient")
ax.set_ylabel("Cluster Label")
ax.set_yticks([])
plt.tight_layout()
plt.show()
```
